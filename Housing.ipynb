```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error


df = pd.read_csv("Housing.csv")

# 2️⃣ حذف القيم الناقصة (لو موجودة)
df = df.dropna()

# 3️⃣ تحويل الأعمدة النصية
df = pd.get_dummies(df, drop_first=True)

# 4️⃣ فصل X و y
X = df.drop("price", axis=1)   # تأكد اسم العمود
y = df["price"]

# 5️⃣ تقسيم البيانات 80%
x_trin, x_test, y_trin, y_test = train_test_split(
    X, y, train_size=0.8, random_state=15
)

# 6️⃣ Scaling (اختياري لكن مفيد مع Polynomial)
scaler = StandardScaler()
x_trin = scaler.fit_transform(x_trin)
x_test = scaler.transform(x_test)

# 7️⃣ Polynomial Features
poly = PolynomialFeatures(degree=2, include_bias=False)

X_train_poly = poly.fit_transform(x_trin)
X_test_poly = poly.transform(x_test)

# 8️⃣ تدريب LinearRegression
model = LinearRegression()
model.fit(X_train_poly, y_trin)

# 9️⃣ التوقع والتقييم
pred = model.predict(X_test_poly)

print("R2 :", r2_score(y_test, pred)*100)
print("MAE:", mean_absolute_error(y_test, pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, pred)))

# dff=pd.DataFrame({
#     "predict":pred,
#     "actual": y_test,
#     "error": pred- y_test
# })

```

```python
from sklearn.metrics import mean_absolute_percentage_error

mape = mean_absolute_percentage_error(y_test, pred)
print(mape*100, "%")
```
